```python
gendermapping ={
    "M":0, 
    "F":1
}
df.Gender= df["Gender"].map(gendermapping)
```

```python
classmapping = {
    "N": 0,
    "P":1,
    "Y":2
}
df.CLASS = df["CLASS"].map(classmapping)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
y=df['CLASS']
X= df.drop("CLASS", axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, stratify=df["CLASS"], random_state=23)
```

```python
# Logistic regresion training
from sklearn.linear_model import LogisticRegression
Regmodel = LogisticRegression()
```

```python
Regmodel.fit(X_train, y_train)
```

```python
Regmodel.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
y_pred = Regmodel.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
display = ConfusionMatrixDisplay(confusion_matrix=cm)
display.plot()
```

```python
from sklearn.metrics import f1_score, accuracy_score, classification_report
print("f1 score", f1_score(y_test, y_pred, average='macro'))
print("Accuracy score", accuracy_score(y_test, y_pred))
print(" Classification report",classification_report(y_test, y_pred))
```

```python
# Resampling 
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import RandomUnderSampler
```

```python
smo = SMOTE()
```

```python
X_resampled, y_resampled = smo.fit_resample(X_train,y_train)
```

```python
sns.countplot(x=y_resampled)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
Resampledmodel = LogisticRegression()
```

```python
Resampledmodel.fit(X_resampled, y_resampled)
```

```python
Resampledmodel.score(X_test, y_test)
y_res= Resampledmodel.predict(X_test)
```

```python
from sklearn.metrics import f1_score, accuracy_score, classification_report
print("f1 score", f1_score(y_test, y_res, average='macro'))
print("Accuracy score", accuracy_score(y_test, y_res))
print(" Classification report",classification_report(y_test, y_res))
```

```python
# Catboost on resampled data
from catboost import CatBoostClassifier
cbs= CatBoostClassifier(random_state = 42, verbose = 100)
cbs.fit(X_resampled, y_resampled)
y_pred_cb = cbs.predict(X_test)
print(accuracy_score(y_test, y_pred_cb))
print(classification_report(y_test, y_pred_cb))
```

```python
cm = confusion_matrix(y_test, y_pred_cb)
display = ConfusionMatrixDisplay(confusion_matrix=cm)
display.plot()
```

```python
# Catboost on unresampled data
from catboost import CatBoostClassifier
cb = CatBoostClassifier(random_state = 42, verbose = 100)
cb.fit(X_train, y_train)
y_pred_cb = cb.predict(X_test)
print(accuracy_score(y_test, y_pred_cb))
print(classification_report(y_test, y_pred_cb))
```

```python
cm = confusion_matrix(y_test, y_pred_cb)
display = ConfusionMatrixDisplay(confusion_matrix=cm)
display.plot()
```
